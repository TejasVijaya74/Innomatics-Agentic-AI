Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes):
    total_likes = 0

    # Calculate total likes using loop
    for like in likes:
        total_likes += like

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

analyze_post_engagement([300, 250, 500])
```

Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0

    for value in rainfall_data:
        total += value

    average = total / len(rainfall_data)

    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

rainfall_checker([70, 80, 65, 75], required_level=70)
```

Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

detect_duplicate_accounts(["user1", "user2", "user1"])
```

Agriculture – Premium Crop Price Filter

```python
def premium_crop_filter(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

premium_crop_filter([1500, 2500, 1800, 3200])
```

System Monitoring – Application Health Checker

```python
def application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

application_health(3)
```

E-Learning – Student Attendance Eligibility System

```python
def attendance_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_eligibility(["P", "P", "A", "P"])
```

Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill: ₹", bill)
    print("Usage Level:", usage)

electricity_bill(260)
```
