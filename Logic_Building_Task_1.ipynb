**1. User Login Check**

```python
#Given Correct Credentials
correct_username = "admin"
correct_password = "1234"

#User Input
username = input("Enter the username: ")
password = input("Enter the password: ")

#Compare between user input and given correct credentials
if username == correct_username and password == correct_password:
  print("Login Successful")
else:
  print("Invalid Credentials")
```

**2. Pass / Fail Analyzer**

```python
#Given marks of students in a list
marks = [45, 78, 90, 33, 60]

#Intial values
passed = 0
failed = 0

for m in marks: #m: marks of a student in a marks list
  if m>=50: # if marks is greater than or equal to 50
  # Adds up if the condition in the if statement satisfies
    passed +=1
  else:
    failed +=1
print("Total Students",len(marks))
print("Total Passed", passed)
print("Total Failed", failed)
```

**3. Simple Data Cleaner**

```python
#Given List
names = [" Alice ", "bob", " CHARLIE "]
#Another new list to add the names, initial it will be null
cleaned_names = []
for name in names:
#strip() method helps to remove whitespaces from the start and the end of the string
#lower() function converts the string to lower case
  cleaned_name = name.strip().lower()
#appends each cleaned name in the cleaned names list
  cleaned_names.append(cleaned_name)
print(cleaned_names)
```

**4. Message Length Analyzer**

```python
#Given message
messages = ["Hi", "Welcome to the platform", "OK"]
for words in messages:
  length = len(words) #finds the length of the each words
  print(f"words: '{words}' | Length: {length}")
# if the lenght is greater the 10 Char it flags a message
  if length > 10 :
    print("Flag : Longer than 10 characters")
```
