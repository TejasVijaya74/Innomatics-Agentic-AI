Smart Parking Lot Management System

```python
def parking_management(capacity, logs):
    """
    Calculates currently parked vehicles,
    peak usage, and alerts if capacity exceeded.
    """

    current = 0
    peak = 0

    # Process each log entry
    for entry in logs:
        if entry == "IN":
            current += 1
        elif entry == "OUT":
            current -= 1

        # Track peak usage
        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    # Capacity check
    if current > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak)

parking_management(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance, traffic, weather):
    """
    Estimates delivery time based on distance,
    traffic level, and weather conditions.
    """

    base_time_per_km = 5  # minutes per km
    eta = distance * base_time_per_km

    # Traffic adjustments
    if traffic.lower() == "high":
        eta += 15
    elif traffic.lower() == "medium":
        eta += 8

    # Weather adjustments
    if weather.lower() == "rainy":
        eta += 10
    elif weather.lower() == "storm":
        eta += 20

    print("Estimated Delivery Time:", int(eta), "minutes")

delivery_time_estimator(8, "High", "Rainy")
```

Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):
    """
    Calculates occupancy percentage and show status.
    """

    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print("Show Status:", status)

theatre_occupancy(200, [1] * 150)
```

Cloud Server Load Classification System

```python
def server_load(cpu_readings):
    """
    Classifies server load based on average CPU usage.
    """

    total = 0

    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    # Classification rules
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Server Status:", status)

server_load([45, 60, 70, 85, 90])
```

Smart Classroom Resource Usage Monitor

```python
def resource_monitor(usage):
    """
    Identifies overused resources.
    Overuse threshold assumed > 8 hours.
    """

    overused = []

    for resource, hours in usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    """
    Controls event registrations,
    prevents overbooking, and manages waitlist.
    """

    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

event_registration(100, 105)
```
