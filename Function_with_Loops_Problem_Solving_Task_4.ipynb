Mobile Recharge Validation System

```python
def validate_recharge():
    """
    Validates mobile recharge requests.
    Conditions:
    - Minimum amount ₹50
    - Must match valid plan
    - Allows retry until valid input is entered
    """

    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Invalid: Minimum recharge is ₹50. Try again.\n")
        elif amount not in valid_plans:
            print("Invalid plan. Available plans:", valid_plans, "\n")
        else:
            print("Recharge successful!")
            break

validate_recharge()
```

Inventory Reorder Alert System

```python
def inventory_alert(inventory):
    """
    Prints Reorder Alert if stock < 15,
    otherwise prints Stock OK.
    """

    for product, quantity in inventory.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock: {quantity})")
        else:
            print(f"{product}: Stock OK (Stock: {quantity})")

inventory = {
        "Rice": 20,
        "Oil": 10,
        "Sugar": 5,
        "Wheat": 25
    }

inventory_alert(inventory)
```

Student Result Processing System

```python
def process_student_result(marks):
    """
    Calculates average marks and returns Pass or Fail.
    Pass condition: average ≥ 50
    """

    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    print(f"Average Marks: {average:.2f}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 55, 70, 45]
result = process_student_result(marks)
print("Result:", result)
```

Cab Fare Estimator with Retry Option

```python
def cab_fare_estimator():
    """
    Calculates cab fare based on distance and peak hours.
    Base fare = ₹50
    ₹12 per km
    Peak hour → 25% extra
    Allows retry using while loop
    """

    while True:
        distance = float(input("Enter distance (km): "))

        base_fare = 50
        fare = base_fare + (12 * distance)

        peak = input("Is it peak hour? (yes/no): ").lower()

        if peak == "yes":
            fare *= 1.25  # Add 25% extra

        print(f"Estimated Fare: ₹{fare:.2f}")

        retry = input("Calculate again? (yes/no): ").lower()
        if retry != "yes":
            break

cab_fare_estimator()
```

Employee Attendance Eligibility Checker

```python
def attendance_eligibility(attendance):
    """
    Calculates attendance percentage.
    Returns 'Eligible' if ≥ 75%, else 'Not Eligible'.
    """

    present_days = 0

    # Count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    print(f"Attendance Percentage: {percentage:.2f}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
eligibility = attendance_eligibility(attendance)
print("Eligibility:", eligibility)
```

Password Strength Checker

```python
def password_strength(password):
    """
    Checks password strength based on rules:
    - Minimum length = 8
    - At least one digit
    - At least one special character (@ # $)
    """

    if len(password) < 8:
        return "Weak Password: Minimum length is 8"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if not has_digit:
        return "Weak Password: Must include a digit"
    if not has_special:
        return "Weak Password: Must include a special character (@ # $)"

    return "Strong Password"

pwd = input("Enter password: ")
print(password_strength(pwd))
```
