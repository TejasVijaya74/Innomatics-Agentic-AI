**1. User Login Check**

```python
correct_username = "admin"
correct_password = "1234"

username = input("Enter the username: ")
password = input("Enter the password: ")

if username == correct_username and password == correct_password:
  print("Login Successful")
else:
  print("Invalid Credentials")
```

**2. Pass / Fail Analyzer**

```python
marks = [45, 78, 90, 33, 60]
passed = 0
failed = 0

for m in marks:
  if m>=50:
    passed +=1
  else:
    failed +=1
print("Total Students",len(marks))
print("Total Passed", passed)
print("Total Failed", failed)
```

**3. Simple Data Cleaner**

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
for name in names:
  cleaned_name = name.strip().lower()
  cleaned_names.append(cleaned_name)
print(cleaned_names)
```

**4. Message Length Analyzer**

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for words in messages:
  length = len(words)
  print(f"words: '{words}' | Length: {length}")

  if length > 10 :
    print("Flag : Longer than 10 characters")
```

**5. Error Message Detector**

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")
print("Total count of ERROR in logs: ", error_count)
```
