**Smart Light Controller**

```python
class SmartLight:
    """
    Represents a smart light that can be turned ON or OFF.
    """
    def __init__(self,name):
      self.name = name
      self.status = "OFF" # Default State

    def turn_on(self):
      self.status = "ON"

    def turn_off(self):
      self.status = "OFF"

    def display_status(self):
      print(f"{self.name} is {self.status}")

light = SmartLight("Kitchen Light")
light.display_status()
light.turn_on()
light.display_status()
```

**Employee ID Card System**

```python
class EmployeeIDCard:
  """
  Stores employee details and displays ID card information.
  """
  def __init__(self, name, emp_id, department):
    self.name = name
    self.emp_id = emp_id
    self.department = department

  def display_card(self):
    print("Employee ID Card")
    print(f"Name: {self.name}")
    print(f"Employee ID: {self.emp_id}")
    print(f"Department: {self.department}")

employee = EmployeeIDCard("John Doe", "12345", "HR")
employee.display_card()
```

**Mobile Contact Record**

```python
class Contact:
    """
    Stores contact name and phone number.
    """
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone Number: {self.phone_number}")

contact = Contact("John Doe", "123-456-7890")
contact.display_contact()
```

**Product Price Tag Generator**

```python
class Product:
    """
    Stores product details and prints a price tag.
    """
    def __init__ (self, name, price):
      self.name = name
      self.price = price

    def display_price_tag(self):
      print("Price Tag")
      print(f"Name: {self.name}")
      print(f"Price: ${self.price}")

product = Product("Laptop", 1000)
product.display_price_tag()
```

**Movie Rating Display System**

```python
class Movie:
    """
    Stores movie name and rating.
    """
    def __init__(self, name, rating):
      self.name = name
      self.rating = rating

    def display_details(self):
      print(f"Movie: {self.name}")
      print(f"Rating: {self.rating}")

movie = Movie("Inception", 9.0)
movie.display_details()
```

**Delivery Address Manager**

```python
class DeliveryDetails:
    """
    Stores customer name and delivery address.
    """
    def __init__(self,customer_name, address):
      self.customer_name = customer_name
      self.address = address

    def display_details(self):
      print("Delivery Details")
      print(f"Customer Name: {self.customer_name}")
      print(f"Address: {self.address}")

details = DeliveryDetails("John Doe", "123 Main St, City, Country")
details.display_details()
```
